# ORDINAL NUMBER ENCODING

In this data is given in such a way so that we can assign ranks (Grading System) to them.
for categorical feature we can only encode them so that they can be easily used as training data on machine learning models.

```python
import datetime
today_date = datetime.datetime.today()
today_date
```

```python
today_date - datetime.timedelta(3)
```

```python
days = [today_date - datetime.timedelta(x) for x in range(0,15)]
```

```python
# converting this data into the data frame by using pandas modules
import pandas as pd 
data = pd.DataFrame(days)
data.columns = ['Day']
```

```python
data.head()
```

```python
data['weekday']=data['Day'].dt.day_name()
data.head()
```

```python
dictionary = {'Monday':1,'Tuesday':2,'Wednesday':3,'Thursday':4,
              'Friday':5,'Saturday':6,'Sunday':7
    
}
```

```python
dictionary
```

```python
data['weekday_ordinal']= data['weekday'].map(dictionary)
```

```python
data.head(10)
```

# TARGET GUIDED ORDINAL ENCODING

```python
# ORDERING THE LABELS ACCORDING TO THE TARGET
# REPLACE THE LABELS BY USING THE JOINT PROBABILITY OF BEING 0 OR 1
```

```python
import pandas as pd 
df = pd.read_csv('train.csv',usecols=['Cabin','Survived'])
```

```python
df.head()
```

```python
df['Cabin'].fillna('Missing',inplace = True)
```

```python
df.head()
```

```python
df['Cabin'] = df['Cabin'].astype(str).str[0]
df.head(10)
# astype is used to convert any existential column to the categorical variables.
```

```python
df.Cabin.unique()
```

```python
df.groupby(['Cabin'])['Survived'].mean()
```

```python
ordinal_labels = df.groupby(['Cabin'])['Survived'].mean().sort_values().index
```

```python
ordinal_labels
```

```python
ordinal_labels2 = {k:i for i,k in enumerate(ordinal_labels,0)}
# enumerate is used to counter the categorical variables(iterable function)
```

```python
ordinal_labels2
```

```python
df['cabin_ordinal_label'] = df['Cabin'].map(ordinal_labels2)
df.head(30)
```

# MEAN ENCODING

```python
mean_ordinal = df.groupby(['Cabin'])['Survived'].mean().to_dict()
```

```python
mean_ordinal
```

```python
df['cabin_mean_ordinal'] = df['Cabin'].map(mean_ordinal)
df.head(30)
```
